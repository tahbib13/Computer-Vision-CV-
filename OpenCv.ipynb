# Video Reading and showing

```python
import cv2

video = cv2.VideoCapture('Comedy.mp4')

while True:
    ret, frame = video.read()

    if not ret:
        break   # End of video

    cv2.imshow('frame', frame)

    # Press q to quit
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```

```python
import cv2

cap = cv2.VideoCapture(1)   # 0 = default webcam

while True:
    ret, frame = cap.read()

    if not ret:
        break

    cv2.imshow("Webcam", frame)

    # Press q to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

# Changing the color space

```python
image = cv2.imread('Broccoli.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("Broccoli", gray_image)
cv2.waitKey(0)
```

# Converting to HSV

A color space to note is HSV. We can create comparisions between image with it using histograms. Lets convert our image to this

```python
import cv2

image = cv2.imread('Broccoli.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
cv2.imshow("Broccoli", gray_image)
cv2.waitKey(0)
```

# Resizing images

Lets see how we can resize image. To do this we need to run the cv2.resize() function.



```python
import cv2

image = cv2.imread('Broccoli.jpg')
small_image = cv2.resize(image, (300, 300)) # resize image
cv2.imshow("Broccoli", small_image)
cv2.waitKey(0)
```

# Saving images

We can save an image by calling the cv2.imwrite() function. It takes the name of the image and the object to save. Lets see the code below

```python
import cv2

image = cv2.imread('Broccoli.jpg')
small_image = cv2.resize(image, (300, 300))
cv2.imwrite("Broccoli_Small.png", small_image)
```

# Basic Blur Example

```python
import cv2

img = cv2.imread("Bird01.jpg")

blur = cv2.blur(img, (15,15))

cv2.imshow("Original", img)
cv2.imshow("Blur Image", blur)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Better Blur (Gaussian Blur)

```python
import cv2

img = cv2.imread("Bird01.jpg")

blur = cv2.GaussianBlur(img, (15,15), 0)

cv2.imshow("Blurred", blur)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
