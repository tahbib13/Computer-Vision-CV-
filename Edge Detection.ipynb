Edge detection is a technique used to identify boundaries of objects in an image.

Edges usually occur where there is a sudden change in pixel intensity (brightness/color).

It helps detect:

Shapes

Object outlines

Corners

Lane lines

Text boundaries

Medical image structures

Why Use Edge Detection?

Examples:

Face detection preprocessing

Self-driving car lane detection

Object counting

Document scanning

Video surveillance

# Canny Edge Detection

```python
import cv2

img = cv2.imread("Basketball.jpg", 0)

edges = cv2.Canny(img, 100, 200)

cv2.imshow("Edges", edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Parameters:

100 = lower threshold

200 = upper threshold

# Sobel Edge Detection

Detects horizontal or vertical edges using gradients.

```python
import cv2
import numpy as np

img = cv2.imread("Basketball.jpg", 0)

sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)

cv2.imshow("Sobel X", sobelx)
cv2.imshow("Sobel Y", sobely)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

dx=1, dy=0 → Vertical edges

dx=0, dy=1 → Horizontal edges

# Laplacian Edge Detection

Detects edges in all directions.

```python
img = cv2.imread("Basketball.jpg", 0)

lap = cv2.Laplacian(img, cv2.CV_64F)

cv2.imshow("Laplacian", lap)
cv2.waitKey(0)
```

# Best Practice Before Edge Detection

```python
img = cv2.imread("Basketball.jpg", 0)

img = cv2.GaussianBlur(img, (5,5), 0)
lap = cv2.Laplacian(img, cv2.CV_64F)

cv2.imshow("Laplacian", lap)
cv2.waitKey(0)
```

# Real-Time Webcam Edge Detection

```python
import cv2

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 100, 200)

    cv2.imshow("Webcam Edge", edges)

    if cv2.waitKey(1) == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
