In Computer Vision (CV), thresholding is an image processing technique used to convert a grayscale image into a binary image (black and white) based on a threshold value.

It helps separate foreground objects from the background.

# Binary Threshold

```python
import cv2

img = cv2.imread('Bear.jpg', 0)

_, th = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

cv2.imshow("Binary", th)
cv2.waitKey(0)
```

# Binary Inverse Threshold

```python
import cv2

img = cv2.imread('Bear.jpg', 0)

_, th = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)

cv2.imshow("Binary", th)
cv2.waitKey(0)
```

# Truncate Threshold

```python
import cv2

img = cv2.imread('Bear.jpg', 0)

_, th = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)

cv2.imshow("Binary", th)
cv2.waitKey(0)
```

```python
import cv2

img = cv2.imread('Bear.jpg', 0)

_, th = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)

cv2.imshow("Binary", th)
cv2.waitKey(0)
```

# To Zero Inverse

```python
import cv2

img = cv2.imread('Bear.jpg', 0)

_, th = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO_INV)

cv2.imshow("Binary", th)
cv2.waitKey(0)
```

# Adaptive Thresholding

```python
import cv2

img = cv2.imread('Bear.jpg', 0)

th = cv2.adaptiveThreshold(
    img,
    255,
    cv2.ADAPTIVE_THRESH_MEAN_C,
    cv2.THRESH_BINARY,
    11,
    2
)

cv2.imshow("Binary", th)
cv2.waitKey(0)
```

# Otsu Thresholding

```python
import cv2

img = cv2.imread('Bear.jpg', 0)

_, th = cv2.threshold(
    img, 0, 255,
    cv2.THRESH_BINARY + cv2.THRESH_OTSU
)

cv2.imshow("Binary", th)
cv2.waitKey(0)
```

# Full Example Showing All Types

```python
import cv2

img = cv2.imread('Bear.jpg',0)

types = [
    cv2.THRESH_BINARY,
    cv2.THRESH_BINARY_INV,
    cv2.THRESH_TRUNC,
    cv2.THRESH_TOZERO,
    cv2.THRESH_TOZERO_INV
]

for i, t in enumerate(types):
    _, th = cv2.threshold(img,127,255,t)
    cv2.imshow(str(i), th)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
