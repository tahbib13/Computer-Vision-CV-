In OpenCV, you can draw shapes and text on images such as:

Line

Rectangle

Circle

Ellipse

Polygon

Text

Useful for:

Bounding boxes in object detection

Face detection marks

Annotation

UI overlays

Tracking systems

# Create Blank Image

```python
import cv2
import numpy as np

img = np.zeros((500,500,3), dtype="uint8")
```

# Draw Line

```python
cv2.line(img, (50,50), (400,50), (0,255,0), 3)
cv2.imshow("Drawing", img)
cv2.waitKey(0)
```

# Draw Rectangle

```python
cv2.rectangle(img, (50,100), (250,250), (255,0,0), 3)
cv2.imshow("Drawing", img)
cv2.waitKey(0)
```

# Filled rectangle:

```python
cv2.rectangle(img, (300,100), (450,250), (0,0,255), -1)
cv2.imshow("Drawing", img)
cv2.waitKey(0)
```

# Draw Circle

```python
cv2.circle(img, (150,350), 60, (0,255,255), 4)
cv2.imshow("Drawing", img)
cv2.waitKey(0)
```

# Filled Circle:

```python
cv2.circle(img, (350,350), 50, (255,255,0), -1)
cv2.imshow("Drawing", img)
cv2.waitKey(0)
```

# Draw Text

```python
cv2.putText(
    img,
    "Tahbib",
    (250,470),
    cv2.FONT_HERSHEY_SIMPLEX,
    1,
    (255,255,255),
    2
)
cv2.imshow("Drawing", img)
cv2.waitKey(0)
```
